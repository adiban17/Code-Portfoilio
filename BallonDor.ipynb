What is the average recovery rate (Performance_misc-Recov) for defenders (DF) compared to midfielders (MF)?

```python
defs=dataset[dataset['pos']=='DF']
avg_rr_def=defs['Performance_misc-Recov'].mean()
avg_rr_def
```

```python
mids=dataset[dataset['pos']=='MF']
avg_rr_mids=mids['Performance_misc-Recov'].mean()
avg_rr_mids
```

```python
comps=avg_rr_def/avg_rr_mids
comps
```

What is the percentage of successful aerial duels (Aerial Duels-Won%) for players above the age of 25?

```python
above_25=dataset[dataset['age']>25]
s_a_d=above_25['Aerial Duels-Won%'].mean()
s_a_d
```

```python
below_25=dataset[dataset['age']<=25]
s_a_d_b_25=below_25['Aerial Duels-Won%'].mean()
s_a_d_b_25
```
